6. Bar Charts

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' with categorical columns
# Replace 'categorical_column_1', 'categorical_column_2', etc. with your actual column names

# Assume you have a DataFrame named 'df' with actual categorical column names
df = pd.DataFrame({
    'categorical_column_1': ['Category_A', 'Category_B', 'Category_A', 'Category_A', 'Category_B'],
    'categorical_column_2': ['Type_X', 'Type_Y', 'Type_X', 'Type_Y', 'Type_X'],
    'categorical_column_3': ['Label_1', 'Label_2', 'Label_1', 'Label_3', 'Label_2'],
})

# Specify key categorical columns to visualize
categorical_columns_to_visualize = ['categorical_column_1', 'categorical_column_2', 'categorical_column_3']

# Specify key categorical columns to visualize
categorical_columns_to_visualize = ['categorical_column_1', 'categorical_column_2']

# Calculate counts for each category in each column
category_counts = df[categorical_columns_to_visualize].apply(pd.value_counts)

# Plot bar charts using Pandas plot
category_counts.T.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')

# Add titles and labels
plt.title('Count of Categorical Variables', fontsize=16)
plt.xlabel('Category')
plt.ylabel('Count')
plt.legend(title='Categorical Variable', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```

8. Pie Charts

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' with categorical columns
# Replace 'car_model', 'manufacturer', etc. with your actual column names
df = pd.DataFrame({
    'car_model': ['Sedan', 'SUV', 'Hatchback', 'Sedan', 'SUV', 'SUV', 'Hatchback', 'Sedan'],
    'manufacturer': ['Toyota', 'Honda', 'Ford', 'Toyota', 'Honda', 'Ford', 'Nissan', 'Toyota'],
    'color': ['Red', 'Blue', 'Black', 'White', 'Red', 'Blue', 'Black', 'White']
})

# Specify the categorical column to visualize
categorical_column_to_visualize = 'car_model'

# Calculate the contribution of each category
category_contribution = df[categorical_column_to_visualize].value_counts()

# Create a pie chart using Pandas plot
category_contribution.plot.pie(autopct='%1.1f%%', startangle=90, figsize=(8, 8))

# Add title
plt.title(f'Contribution of {categorical_column_to_visualize}', fontsize=16)

# Show the plot
plt.show()
```
