```python
# Drop rows with missing values in any column
df = df.dropna()
```

```python
# Replace missing values in 'your_column' with the mean of that column
if 'your_column' in df.columns:
    df['your_column'] = df['your_column'].fillna(df['your_column'].mean())
```

```python
# Display summary statistics
summary_stats = df.describe()
print("Summary Statistics:\n", summary_stats)
```

```python
df['New_Column'] = [1] * len(df)
sns.histplot(df['New_Column'])
plt.title('Distribution of New_Column')
plt.show()
```

```python
# Choose and implement a strategy (e.g., replace with mean)
df.fillna(df.mean(), inplace=True)

# Visual inspection (you may need to install additional libraries)
sns.boxplot(data=df)  # Example of a boxplot for outlier detection
```

7. Seasonal Decomposition

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Load the dataset (replace 'path/to/your/IndianCarsDataset.csv' with the actual file path)
df = pd.read_csv('/content/train.csv')
```

```python
# Check the columns in your DataFrame
print(df.columns)

# Assuming the date column is present and named 'Date', change the index
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
else:
    print("Error: 'Date' column not found in the DataFrame.")

# Display the first few rows of the data
df.head()
```

```python
# Replace the following line with your actual data
# data with a hypothetical time series
time_series_data = pd.Series(np.sin(np.linspace(0, 100, 365)) + np.random.normal(0, 0.1, 365), name='Value')

# Plot the original time series data
plt.figure(figsize=(12, 6))
plt.plot(time_series_data)
plt.title('Original Time Series Data')
plt.show()
```

decompose the time series into its components

```python
# Decompose the time series data into trend, seasonality, and residuals
result = seasonal_decompose(time_series_data, model='additive', period=1)  # You can adjust the period parameter based on the seasonality of your data

# Plot the decomposed components
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(result.observed)
plt.title('Original Time Series Data')

plt.subplot(4, 1, 2)
plt.plot(result.trend)
plt.title('Trend Component')

plt.subplot(4, 1, 3)
plt.plot(result.seasonal)
plt.title('Seasonal Component')

plt.subplot(4, 1, 4)
plt.plot(result.resid)
plt.title('Residual Component')

plt.tight_layout()
plt.show()
```

10. Sales Variation Over Time

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset (replace 'path/to/your/IndianCarsDataset.csv' with the actual file path)
df = pd.read_csv('/content/train.csv')
```

```python
# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_sales = [10000, 12000, 15000, 11000, 13000, 16000, 18000, 20000, 19000, 22000, 25000, 23000]

quarters = ['Q1', 'Q2', 'Q3', 'Q4']
quarterly_sales = [35000, 48000, 60000, 70000]

years = ['2018', '2019', '2020', '2021', '2022', '2023']
yearly_sales = [180000, 200000, 220000, 250000, 280000, 300000]

# Monthly Sales Line Chart
plt.figure(figsize=(10, 6))
sns.lineplot(x=months, y=monthly_sales, marker='o', color='b')
plt.title('Monthly Sales Variation')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.show()

# Quarterly Sales Line Chart
plt.figure(figsize=(10, 6))
sns.lineplot(x=quarters, y=quarterly_sales, marker='o', color='g')
plt.title('Quarterly Sales Variation')
plt.xlabel('Quarter')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.show()

# Yearly Sales Line Chart
plt.figure(figsize=(10, 6))
sns.lineplot(x=years, y=yearly_sales, marker='o', color='r')
plt.title('Yearly Sales Variation')
plt.xlabel('Year')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.show()
```
